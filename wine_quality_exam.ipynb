```python
ip=data.drop('quality',axis=1)
op=data['quality']
ip
```

```python
op
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.1)
```

```python
#standar scalar 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp = alg1.predict(x_test)
```

```python
yp
```

```python
data1=pd.DataFrame({'y':list(y_test),'prediction':yp})
data1
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
from sklearn.metrics import accuracy_score,recall_score
print("Accuracy Score:", accuracy_score(y_test, yp))
print("Recall",recall_score(y_test,yp, average='macro') )
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
alg2 = KNeighborsClassifier(n_neighbors = 5)
alg2.fit(x_train, y_train)
```

```python
yp2 = alg2.predict(x_test)
yp2
```

```python
from sklearn.metrics import accuracy_score, recall_score
print("Accuracy Score :", accuracy_score(y_test, yp2))
print("Recall :",recall_score(y_test,yp, average='macro') )
```

```python
from sklearn.tree import DecisionTreeClassifier
alg3 = DecisionTreeClassifier(criterion='entropy')
alg3 = alg3.fit(x_train,y_train)
yp3= alg3.predict(x_test)
#Decision tree
```

```python
from sklearn import metrics
accuracy2 = metrics.accuracy_score(y_test, yp3)
recall2 = metrics.recall_score(y_test, yp3, average='macro')
print("Accuracy Score:", accuracy2)
print("Recall:", recall2)
```

```python
from sklearn.ensemble import RandomForestClassifier
alg4=RandomForestClassifier(criterion='entropy',random_state=0)
alg4.fit(x_train,y_train)
yp4=alg4.predict(x_test)
```

```python
accuracy3=metrics.accuracy_score(y_test,yp4)
recall3=metrics.recall_score(y_test,yp4,average='macro')
print("Accuracy_Score :",accuracy3)
print("Recall :",recall3)
```
